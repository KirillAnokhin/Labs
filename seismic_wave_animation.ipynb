```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns; sns.set_theme()
 
dim = 1000
v = 5000
 
arr_t = np.array([t for t in range(dim)], dtype='f')
arr_x = np.array([x for x in range(dim)], dtype='f')
 
arr_t=arr_t*0.001
 
ylist= np.array([t for t in range(dim)], dtype='f')
 
# print(arr_t)
 
arr_u = np.zeros((dim, dim))
arr_u[0][0] = 0
for t in range(95, 158, 1):
  arr_u[0][t] = -math.sin(100*arr_t[t]) 
for t in range(1, dim, 1):
  for x in range(1, t, 1):
    arr_u[x][t] = arr_u[0][round(t-x)]
    
for t in range(1, dim, 1):
    ylist[t]=arr_u[500][t];
 
plt.title('Seismic wave intensity')
plt.xlabel('t (s)')
plt.ylabel('Intensity')
plt.plot(arr_t,ylist,'m.')
plt.gcf().set_size_inches(20, 16)
plt.show()
```
